```python
df["Name"].fillna(df["Name"].mode()[0], inplace = True )
```

```python
df["Team"].fillna(df["Team"].mode()[0], inplace = True )
```

```python
df["Position"].fillna(df["Position"].mode()[0], inplace = True )
```

```python
df["Height"].fillna(df["Height"].mode()[0], inplace = True )
```

```python
df["College"].fillna(df["College"].mode()[0], inplace = True )
```

```python
df["Number"].fillna(df["Number"].mean(), inplace = True )
```

```python
df["Age"].fillna(df["Age"].mean(), inplace = True )
```

```python
df["Weight"].fillna(df["Weight"].mean(), inplace = True )
```

```python
df["Salary"].fillna(df["Salary"].mean(), inplace = True )
```

```python
numeric_cols = ['Number', 'Age']
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
df
```

sns.boxplot(df)

```python
sns.set(style="whitegrid")
numeric_df = df.select_dtypes(include=['float64', 'int64'])


plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Check if your dataset has a 'Number' column for jersey numbers
if 'Number' in df.columns:
    # Calculate average salary for each jersey number
    number_avg_salary = df.groupby('Number')['Salary'].mean()
    
    # Find the jersey number with the highest average salary
    lucky_number = number_avg_salary.idxmax()
    lucky_number_avg_salary = number_avg_salary.max()
    
    print("Lucky Jersey Number:", lucky_number)
    print("Highest Average Salary for that Number:", lucky_number_avg_salary)
else:
    print("Dataset does not contain a 'Number' column for jersey numbers.")
```

```python
sns.pairplot(df, vars=['Number', 'Age', 'Weight', 'Salary'])
plt.suptitle('Pairplot of Numeric Features by Position')
plt.show()
```

```python
sns.boxplot(df)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Position', y='Salary')
plt.title('Salary Distribution by Position')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
team_counts = df['Team'].value_counts()
sns.barplot(x=team_counts.index, color = "red" ,y=team_counts.values)
plt.xticks(rotation=90)
plt.title('Number of Players per Team')
plt.xlabel('Team')
plt.ylabel('Number of Players')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
age_salary = df.groupby('Age')['Salary'].mean().sort_index()
plt.plot(age_salary.index, age_salary.values, color = "b" , marker='o', linestyle='--')
plt.title('Average Salary by Age')
plt.xlabel('Age')
plt.ylabel('Average Salary')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7, 7))
position_counts = df['Position'].value_counts()
plt.pie(position_counts, labels=position_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Player Distribution by Position')
plt.axis('equal')  
plt.show()
```

```python
top_paid = df[['Name', 'Salary']].sort_values(by='Salary', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(data=top_paid, x='Salary', y='Name', palette='viridis')
plt.title('Top 10 Highest Paid NBA Players')
plt.xlabel('Salary')
plt.ylabel('Player Name')
plt.show()
```
